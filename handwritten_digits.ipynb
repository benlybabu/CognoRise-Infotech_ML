```python
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential #initialisation
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPool2D
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
```

```python
#loading data
(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

```python
#reshape the data
x_train=x_train.reshape(60000,28,28,1)   # 1 because it is black and white data
x_test=x_test.reshape(10000,28,28,1)
x_train.shape
```

```python
#Normalize the pixcel value
x_train=x_train/255
x_test=x_test/255
```

```python
#defining model
model=Sequential()
#adding convolutional layer
model.add(Conv2D(32,(3,3),activation='relu',input_shape=(28,28,1)))
#adding pooling layer
model.add(MaxPool2D(2,2))
#flatten
model.add(Flatten())
#fully coonnected
model.add(Dense(100,activation='relu'))  #hidden layer
model.add(Dense(10,activation='softmax'))  #output layer
```

```python
#compiling the model
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.fit(x_train,y_train,batch_size=6000,epochs=10)
```
